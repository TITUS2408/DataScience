```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data1=np.array([60,68,70,72,75,70])
```

```python
#Create range of x values for the plot

x=np.arange(len(data1))
```

```python
#Calculate meand and standard deviation
mean= np.mean(data1)
std_dev=np.std(data1)
```

```python
#plot the data
plt.figure(figsize=(10,6))
plt.plot(x,data1,marker='o',linestyle='-',label='data')
#plot horizontal lines for mean and standard deviation
plt.axhline(mean,color='r',linestyle='--',linewidth=0.9,label='Mean')


# Add legend to print the labels
plt.legend()

```
